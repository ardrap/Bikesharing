```python
df=df.rename(columns={'weathersit':'weather',
                      'yr':'year',
                      'mnth':'month',
                      'hr':'hour',
                      'hum':'humidity',
                      'cnt':'count'})
df.head()
```

```python
df=df.drop(columns=['instant','dteday','year'])
df.head()
```

```python
cols=['season','month','hour','holiday','weekday','workingday','weather']

for col in cols:
  df[col]=df[col].astype('category')
df.info()
```

```python
pd.get_dummies(df['season'],prefix='season',drop_first=True)
```

```python
df_oh=df

def one_hot_encoding(data, column):
    data=pd.concat([data,pd.get_dummies(data[column],prefix=column,drop_first=True)],axis=1)
    data=data.drop([column],axis=1)
    return data

cols=['season','month','hour','holiday','weekday','workingday','weather']

for col in cols:
  df_oh=one_hot_encoding(df_oh,col)
df_oh.head()
```

```python
X=df_oh.drop(columns=['atemp','windspeed','casual','registered','count'],axis=1)
y=df_oh['count']
```

```python
# model_selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
# # pre-proceesing/transform

# from sklearn.preprocessing import StandardScaler
# scaler=StandardScaler()
# scaler.fit(x_train)

# x_train=scaler.transform(x_train)
# print(x_train)

# x_test=scaler.transform(x_test)
# print(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lreg= LinearRegression()
lreg.fit(x_train,y_train)
```

```python
y_pred=lreg.predict(x_test)
y_pred
```

```python
# Mean Absolute Error
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
print("Mean Absolute Error is",mean_absolute_error(y_test,y_pred))
# Mean Squared Error
from sklearn.metrics import mean_squared_error,mean_squared_log_error
print("MSE",mean_squared_error(y_test,y_pred))
# Root Mean Squared
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
# variance score
print("Variance score: ",lreg.score(x_test,y_test))
# R2 score
print("R2 score: ",r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier(criterion='entropy')
dec.fit(x_train,y_train)

y_pred_DT=dec.predict(x_test)
y_pred_DT

```

```python
# Mean Absolute Error
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
print("Mean Absolute Error is",mean_absolute_error(y_test,y_pred_DT))
# Mean Squared Error
from sklearn.metrics import mean_squared_error,mean_squared_log_error
print("MSE",mean_squared_error(y_test,y_pred_DT))
# Root Mean Squared
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred_DT)))
# variance score
print("Variance score: ",dec.score(x_test,y_test))
# R2 score
print("R2 score: ",r2_score(y_test,y_pred_DT))
```

```python
from sklearn.ensemble import RandomForestRegressor
RFreg=RandomForestRegressor(n_estimators=3)
RFreg.fit(x_train,y_train)
```

```python
y_pred_RF=RFreg.predict(x_test)
y_pred_RF
```

```python
# Mean Absolute Error
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
print("Mean Absolute Error is",mean_absolute_error(y_test,y_pred_RF))
# Mean Squared Error
from sklearn.metrics import mean_squared_error,mean_squared_log_error
print("MSE",mean_squared_error(y_test,y_pred_RF))
# Root Mean Squared
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred_RF)))
# variance score
print("Variance score: ",RFreg.score(x_test,y_test))
# R2 score
print("R2 score: ",r2_score(y_test,y_pred_RF))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

models=[LinearRegression(),
        DecisionTreeRegressor(),
        RandomForestRegressor()]
```

```python
from sklearn import model_selection
```

```python
def train(model):
    kfold = model_selection.KFold(n_splits=5)
    pred = model_selection.cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')
    cv_score = pred.mean()
    print('Model:',model)
    print('CV score:', abs(cv_score))
```

```python
for model in models:
  train(model)
```
